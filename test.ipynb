```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
```

# 1.Tien Xu Ly Anh

```python
# Doc anh tu thu muc
data_path = "/home/nguyenngocduong/Desktop/face_recognition_project/data"
images = []
labels = []
for person_name in sorted(os.listdir(data_path)):
    person_folder = os.path.join(data_path, person_name)
    if os.path.isdir(person_folder):
        for image in sorted(os.listdir(person_folder)):
            image_path = os.path.join(person_folder, image)
            img = cv2.imread(image_path)
            if img is None:
                continue
            images.append(img)
            labels.append(person_name)
print(f"Tổng số ảnh: {len(images)}")
print(f"Số class: {len(set(labels))}")
```

```python
cv2.imshow("Sample Image", images[0])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Áp dụng Face Detection để lấy gương mặt

```python
from mtcnn import MTCNN
#detect truoc roi moi resize anh
detector = MTCNN()
faces_data = []
labels_data = []
for image, label in zip(images, labels):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    faces = detector.detect_faces(image_rgb)
    for face in faces:
        x,y, width, height = face['box']
        face_crop = image[y:y+height, x: x + width]
        face_crop = cv2.resize(face_crop, (160, 160))
        faces_data.append(face_crop)
        labels_data.append(label)
print(f"Tổng số gương mặt: {len(faces_data)}")
```

```python
import csv
# Luu anh bang file csv
output_path = "/home/nguyenngocduong/Desktop/face_recognition_project/Face_Detection/faces_data.csv"
if not os.path.exists(output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
with open(output_path, 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['label', 'image'])
    for label, face in zip(labels_data, faces_data):
        face_flat = face.flatten()
        writer.writerow([label, np.array(face_flat)])
```

```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionResNetV2
from tensorflow.keras.applications.inception_resnet_v2 import preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np

print("TensorFlow version:", tf.__version__)
```

```python
model = InceptionResNetV2(weights="imagenet", include_top=False, pooling='avg')
img = image.load_img("data/Dat/pic12.jpg", target_size=(299, 299))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)  # Chuẩn hóa theo style Inception
embedding = model.predict(x)
print("Embedding shape:", embedding.shape)
```
